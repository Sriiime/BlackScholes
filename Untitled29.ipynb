```python
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
# Define the Black-Scholes model
def black_scholes(S, K, T, r, sigma, option_type):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    if option_type == 'call':
        return S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    elif option_type == 'put':
        return K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
```

```python
# Calculate option prices using Black-Scholes model
data['bs_option_price'] = data.apply(lambda row: black_scholes(row['underlying_asset_price'],
                                                               row['strike_price'],
                                                               row['time_to_maturity'],
                                                               row['risk_free_interest_rate'],
                                                               row['volatility'],
                                                               row['option_type']), axis=1)
```

```python
# Calculate option prices using Binomial model (for comparison)
def binomial_model(S, K, T, r, sigma, option_type, n=100):
    dt = T / n
    u = np.exp(sigma * np.sqrt(dt))
    d = 1 / u
    p = (np.exp(r * dt) - d) / (u - d)
    option_prices = np.zeros(n+1)
    for i in range(n+1):
        option_prices[i] = max(0, (S * (u**i) * (d**(n-i)) - K) if option_type == 'call' else (K - S * (u**i) * (d**(n-i))))
    for t in range(n-1, -1, -1):
        for i in range(t+1):
            option_prices[i] = np.exp(-r * dt) * (p * option_prices[i+1] + (1-p) * option_prices[i])
    return option_prices[0]

data['binomial_option_price'] = data.apply(lambda row: binomial_model(row['underlying_asset_price'],
                                                                       row['strike_price'],
                                                                       row['time_to_maturity'],
                                                                       row['risk_free_interest_rate'],
                                                                       row['volatility'],
                                                                       row['option_type']), axis=1)
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_option_prices(data):
    fig, ax = plt.subplots(figsize=(10,6))
    ax.plot(data['underlying_asset_price'], data['bs_option_price'], label='Black-Scholes', color='orange', linestyle='-', linewidth=2)
    ax.plot(data['underlying_asset_price'], data['binomial_option_price'], label='Binomial', color='blue', linestyle='--', linewidth=2)

    ax.set_title('Option Pricing Models', fontsize=16, fontweight='bold')
    ax.set_xlabel('Underlying Asset Price ($)', fontsize=14)
    ax.set_ylabel('Option Price ($)', fontsize=14)
    ax.grid(True)
    ax.legend(loc='upper left', fontsize=12)

    plt.tight_layout()
    plt.show()

plot_option_prices(data)

```
